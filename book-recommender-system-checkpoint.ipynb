## Popularity based recommender system

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','num_ratings','avg_rating','Image-URL-M']]
```

```python
popular_df['Image-URL-M'][0]
```

## collaborative filtering based recommendation system


```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
rated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(rated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
rated_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(rated_books)]
```

```python
final = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final.fillna(0,inplace=True)
```

```python
final
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final)
```

```python
similarity_score
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(final.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == final.index[i[0]]]
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.append(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data 
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```

```python
recommend('A Bend in the Road')
```

```python
recommend('Zoya')
```
